## Caminante aleatorio

Crearemos una clase *CaminanteAleatorio()* que tomará decisiones azarosas sobre la dirección que debe tomar el caminante, por tanto la la clase necesita de tres atributos, una variable para almacenar el numero de puntos del recorrido y dos listas para lamacenar los valores de las coordenadas (x,y) de cada punto del recorrido.

Ademas la clase *CaminanteAleatorio()* necesita el metodo **__init()** y **fill_walk()** el cual calcula los puntos de la caminata

```python
#Llamamos al modulo random para decidir que movimiento hacer
from random import choice

class CaminanteAleatorio():
    def __init__(self, num_puntos = 5000):
        #Iniciamos los atributos del caminante
        self.num_puntos = num_puntos
        #Las caminatas cominezan en cero en ambas direcciones
        self.valores_x = [0]
        self.valores_y = [0]
    
    #metodo fill_walk() para llenar la caminata de puntos
    def fill_walk(self):
        #seguiremos dando pasos hasta que la caminata alcance la longitud deseada
        while len(self.valores_x) < self.num_puntos:
            #Caso x
            direccion_x = choice([1,-1])
            distancia_x = choice([0,1,2,3,4])
            paso_x = direccion_x * distancia_x

            #Caso y
            direccion_y = choice([1,-1])
            distancia_y = choice([0,1,2,3,4])
            paso_y = direccion_y * distancia_y

            #Filtramos los movimientos que no van a ningun lado
            if paso_x == 0 and paso_y == 0:
                continue

            #Calculamos la nueva posicion
            x = self.valores_x[-1] + paso_x
            y = self.valores_y[-1] + paso_y

            self.valores_x.append(x)
            self.valores_y.append(y)
        
```

Ahora graficamos la trayectoria del caminante diferenciando el camino que sigue desde el inicio hasta el final, mediante un mapa de colores:)

Para apreciar de mejor manera los puntos, les quitamos el contorno y marcamos el punto incial y final de la caminata

```python
import matplotlib.pyplot as plt

#Hacemos una instancia de la clase y accedemos a la funcion fill_walk()  ;v
rw = CaminanteAleatorio()
rw.fill_walk()

#Ahora imprimimos los puntos del caminante
plt.style.use('classic')
fig, ax = plt.subplots()
puntos_n = range(rw.num_puntos)
ax.scatter(rw.valores_x, rw.valores_y,c=puntos_n, cmap=plt.cm.Blues, edgecolors='none', s=15)

#Para destacar el punto inicial, llamamos nuevamente ax.scatter() los valores
#iniciales corresponden a la cordenada 0,0 ;v
ax.scatter(0,0, c='green', edgecolors='none',s=100)
#Para el punto final hacemos algo similar, solo que llamamos al ultimo punto
# de la lista valores_x[-1] y valores_y[-1]
ax.scatter(rw.valores_x[-1], rw.valores_y[-1], c='red', edgecolors='none', s=100)

plt.show()
```
